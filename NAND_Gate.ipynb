```python
import turtle

# create a turtle object
t = turtle.Turtle()

# font style and size
gate_name = ("Arial", 18, "bold")
title=("Arial", 16, "bold")
font = ("Arial", 14, "normal")

# draw the NAND gate
def draw_NAND():

    #title - NAND Gate
    t.penup()
    t.goto(-40, 230)
    t.pendown()
    t.write("NAND Gate", font=gate_name)

    #Diagram
    t.penup()
    t.goto(-150,170) 
    t.pendown()

    t.write("A", font=font) 
    t.penup()
    t.forward(20)
    t.pendown()

    t.forward(80)
    t.right(90)
    t.forward(60)
    t.backward(20)
    t.right(90)
    t.forward(80)

    t.penup()
    t.forward(20)
    t.pendown()
    t.write("B", font=font) 

    t.penup()
    t.backward(100)
    t.pendown()

    t.right(90)
    t.forward(60)
    t.right(90)
    t.forward(80)
    t.circle(-40,180)
    t.forward(80)

    t.penup()
    t.right(180)
    t.forward(80)
    t.circle(40,90)
    t.pendown()
    t.circle(-10,360)
    t.penup()
    t.right(90) 
    t.forward(20) 
    t.pendown()
    t.forward(80)

    t.penup()
    t.forward(10)
    t.pendown()
    t.write("C", font=font)

    # Truth Table
    t.penup()
    t.goto(-120, 40)
    t.pendown()
    t.write("Truth Table:", font=title)

    # draw the table header
    t.penup()
    t.goto(-120, 10)
    t.pendown()
    t.write("Input A", font=font)
    t.penup()
    t.goto(-40, 10)
    t.pendown()
    t.write("Input B", font=font)
    t.penup()
    t.goto(40, 10)
    t.pendown()
    t.write("Output C", font=font)

# draw the table content
    t.penup()
    t.goto(-120, -10)
    t.pendown()
    t.write("     0", font=font)
    t.penup()
    t.goto(-40, -10)
    t.pendown()
    t.write("     0", font=font)
    t.penup()
    t.goto(40, -10)
    t.pendown()
    t.write("     1", font=font)

    t.penup()
    t.goto(-120, -30)
    t.pendown()
    t.write("     0", font=font)
    t.penup()
    t.goto(-40, -30)
    t.pendown()
    t.write("     1", font=font)
    t.penup()
    t.goto(40, -30)
    t.pendown()
    t.write("     1", font=font)

    t.penup()
    t.goto(-120, -50)
    t.pendown()
    t.write("     1", font=font)
    t.penup()
    t.goto(-40, -50)
    t.pendown()
    t.write("     0", font=font)
    t.penup()
    t.goto(40, -50)
    t.pendown()
    t.write("     1", font=font)

    t.penup()
    t.goto(-120, -70)
    t.pendown()
    t.write("     1", font=font)
    t.penup()
    t.goto(-40, -70)
    t.pendown()
    t.write("     1", font=font)
    t.penup()
    t.goto(40, -70)
    t.pendown()
    t.write("     0", font=font)

    #hide the turtle and wait for the user to close the window
    t.hideturtle()
    turtle.done()
    
    

    
#output
draw_NAND()
 


```
