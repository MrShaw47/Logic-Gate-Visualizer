```python
# draw the AND gate
def draw_AND():
    # create a turtle object
    t = turtle.Turtle()
    t.screen.clear()

    # font style and size
    gate_name = ("Arial", 18, "bold")
    title=("Arial", 16, "bold")
    font = ("Arial", 14, "normal")
    
    t.clear()
    
    #title - AND Gate
    t.penup()
    t.goto(-40, 230)
    t.pendown()
    t.write("AND Gate", font=gate_name)

    #Diagram
    t.penup()
    t.goto(-150,170) 
    t.pendown()

    t.write("A", font=font) 
    t.penup()
    t.forward(20)
    t.pendown()

    t.forward(80)
    t.right(90)
    t.forward(60)
    t.backward(20)
    t.right(90)
    t.forward(80)

    t.penup()
    t.forward(20)
    t.pendown()
    t.write("B", font=font) 

    t.penup()
    t.backward(100)
    t.pendown()

    t.right(90)
    t.forward(60)
    t.right(90)
    t.forward(80)
    t.circle(-40,180)
    t.forward(80)

    t.penup()
    t.right(180)
    t.forward(80)
    t.circle(40,90)
    t.right(90)  
    t.pendown()
    t.forward(80)

    t.penup()
    t.forward(10)
    t.pendown()
    t.write("C", font=font)

    # Truth Table
    t.penup()
    t.goto(-120, 40)
    t.pendown()
    t.write("Truth Table:", font=title)

    # draw the table header
    t.penup()
    t.goto(-120, 10)
    t.pendown()
    t.write("Input A", font=font)
    t.penup()
    t.goto(-40, 10)
    t.pendown()
    t.write("Input B", font=font)
    t.penup()
    t.goto(40, 10)
    t.pendown()
    t.write("Output C", font=font)

# draw the table content
    t.penup()
    t.goto(-120, -10)
    t.pendown()
    t.write("     0", font=font)
    t.penup()
    t.goto(-40, -10)
    t.pendown()
    t.write("     0", font=font)
    t.penup()
    t.goto(40, -10)
    t.pendown()
    t.write("     0", font=font)

    t.penup()
    t.goto(-120, -30)
    t.pendown()
    t.write("     0", font=font)
    t.penup()
    t.goto(-40, -30)
    t.pendown()
    t.write("     1", font=font)
    t.penup()
    t.goto(40, -30)
    t.pendown()
    t.write("     0", font=font)

    t.penup()
    t.goto(-120, -50)
    t.pendown()
    t.write("     1", font=font)
    t.penup()
    t.goto(-40, -50)
    t.pendown()
    t.write("     0", font=font)
    t.penup()
    t.goto(40, -50)
    t.pendown()
    t.write("     0", font=font)

    t.penup()
    t.goto(-120, -70)
    t.pendown()
    t.write("     1", font=font)
    t.penup()
    t.goto(-40, -70)
    t.pendown()
    t.write("     1", font=font)
    t.penup()
    t.goto(40, -70)
    t.pendown()
    t.write("     1", font=font)

    # hide the turtle and wait for the user to close the window
    t.hideturtle()
    

```

```python
def and_gate(a, b):
    return a & b


def or_gate(a, b):
    return a | b


def not_gate(a):
    return not a


def nand_gate(a, b):
    if a == 1 and b == 1:
        return 0
    else:
        return 1


def nor_gate(a, b):
    if a == 0 and b == 0:
        return 1
    else:
        return 0


def xor_gate(a, b):
    if a == b:
        return 0
    else:
        return 1



```
